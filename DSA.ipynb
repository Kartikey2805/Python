```python
### Exercise
# avg temperatures
import numpy as np
temperatures = np.array([0,0,0], dtype=int)
for i in range(3):
    temp = input(f"What is temperature of {i+1} day? : ")
    temperatures[i] = temp
print(f"Avg Temperature of 3 days is {temperatures.mean()}")
days = len([x for x in temperatures if x > 15])
print(days)
```
